```python
import re

# Define a function to extract the numeric part from the string
def extract_numeric(string):
    # Use regular expression to find numeric part in the string
    match = re.search(r'\d+\.\d+', string)
    if match:
        return float(match.group())
    else:
        return None  # Return None if no numeric part found

# Apply the function to the 'exactPrice' column
data['exactPrice'] = data['exactPrice'].apply(lambda x: extract_numeric(x))
```

```python
#workign on colors
def colors_conversion(df):
    new_rows = []
    for index, row in df.iterrows():
        colors = row['colors']
        if len(colors) > 1:
            for color in colors:
                new_row = row.copy()
                new_row['colors'] = color
                new_rows.append(new_row)
        else:
            new_rows.append(row)
    return pd.DataFrame(new_rows)

data = colors_conversion(data)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Download WordNet corpus if not already downloaded
nltk.download('wordnet')

# Initialize the WordNet lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize the categories in the women_shoe_types list
women_shoe_types_lemmatized = [lemmatizer.lemmatize(word) for word in women_shoe_types]

# Lemmatize the categories in the men_shoes_types list
men_shoes_types_lemmatized = [lemmatizer.lemmatize(word) for word in men_shoes_types]

men_shoes_types_lemmatized
```

```python
# prompt: write me a code to remove all the punctuations and numerical values from a string

import re

def remove_punc_and_num(strings):
    cleaned_strings = []
    for string in strings:
        # Remove punctuations and numerical values using regular expression
        cleaned_string = re.sub(r'[^a-zA-Z\s]', '', string)
        cleaned_strings.append(cleaned_string)
    return cleaned_strings
```

```python
for index, row in data.iterrows():
    name = row['name']
    words = name.lower().split()
    words = remove_punc_and_num(words)
    words = [lemmatizer.lemmatize(word) for word in words]

    category_found = False
    category = None

    if row['gender'] == 'women':
        for word in words:
            if word in women_shoe_types:
                category = word
                category_found = True
                break
    else:
        for word in words:
            if word in men_shoes_types:
                category = word
                category_found = True
                break

    # Assign the category only if found, else keep it as None
    data.at[index, 'category'] = category if category_found else None
```

```python
import numpy as np

# Initialize an empty dictionary to store review counts for each category
review_count_by_category = {}

# Replace NaN values in the "reviews" column with 0
data["reviews"] = data["reviews"].fillna(0)

# Iterate through each row in the DataFrame
for index, row in data.iterrows():
    category = row["category"]
    # If category is not already in the dictionary, add it with a review count of 0
    if category not in review_count_by_category:
        review_count_by_category[category] = 0
    # Add the number of reviews of the current row to the corresponding category
    review_count_by_category[category] += row["reviews"]

# Sort the dictionary items by review count in descending order
sorted_categories = {k: v for k, v in sorted(review_count_by_category.items(), key=lambda x: x[1], reverse=True)}

# Print the sorted dictionary
print(sorted_categories)
```

```python
# import pandas as pd
import pandas as pd

class Prediction:
    def __init__(self, data):
        self.data = data

    def database_filtering(self, **kwargs):
        filtered_data = self.data
        for criterion, value in kwargs.items():
            if value is not None:
                filtered_data = filtered_data[filtered_data[criterion] == value]
        return filtered_data

    def selection(self, recommendation_number, filtered_data, category_counts):
        if filtered_data.shape[0] <= recommendation_number:
            return filtered_data
        else:
            # Normalize category counts to get proportions
            total_products = sum(category_counts.values())
            category_proportions = {category: count / total_products for category, count in category_counts.items()}

            # Calculate number of recommendations for each category
            category_recommendations = {category: int(proportion * recommendation_number) for category, proportion in category_proportions.items()}

            # Sample items from each category based on recommendations
            recommended_items = pd.DataFrame()
            for category, count in category_recommendations.items():
                category_data = filtered_data[filtered_data['category'] == category]
                sampled_data = category_data.sample(min(count, len(category_data)))
                recommended_items = pd.concat([recommended_items, sampled_data])


            # Filter out rows from filtered_data where all categories from sampled_data are present
            recommended_categories = set(recommended_items['category'])
            for i, r in filtered_data.iterrows():
                if r['category'] in recommended_categories:
                    filtered_data.drop(i, inplace=True)


            # Check if the number of recommended items is less than recommendation_number
            if recommended_items.shape[0] != recommendation_number:
                diff = recommendation_number - recommended_items.shape[0]
                # Add diff number of products in recommended items by random sampling from filtered data
                additional_data = filtered_data.sample(n=diff, replace=False)
                recommended_items = pd.concat([recommended_items, additional_data])



            return recommended_items

```

```python

# Create an instance of the Prediction class
a = Prediction(data)

# Get user input for filter parameters
gender = input("Enter gender (leave blank if not filtering by gender): ")
price_range = input("Enter price range (leave blank if not filtering by price range): ")
size = input("Enter size (leave blank if not filtering by size): ")


# Convert empty strings to None
gender = None if gender == '' else gender
price_range = None if price_range == '' else price_range
size = None if size == '' else size


# Filter the dataset based on user input
filtered_data = a.database_filtering(gender=gender, price_range=price_range, sizeDetails=size)

# Get user input for recommendation parameters
recommendation_number = int(input("Enter number of recommendations: "))

# Get recommendations based on filtered data
recommendations = a.selection(recommendation_number,filtered_data,sorted_categories)
# print(recommendations1)
```
